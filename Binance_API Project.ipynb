```python
url = 'https://api.binance.com/api/v3/trades'
```

```python
params = {
    'symbol': 'BTCUSDT',
    'limit': 100
}
```

```python
data = requests.get(url, params, timeout=10)
```

```python
raw_data = data.json()
```

```python
df = pd.DataFrame(raw_data)
```
